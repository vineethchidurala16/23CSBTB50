```python
people = ['kiran','arun','vijay','varun']
age =[25,30,35,40]
height =[145,151,165,173]
weight=[45,55,65,75]
```

```python
import matplotlib.pyplot as plt
# Line Plot for Age vs. Height
plt.figure(figsize=(8, 6))
plt.plot(age, height, marker='o', linestyle='-')
plt.title('Age vs. Height')
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(age, weight, marker='o', linestyle='-')
plt.title('Age vs. Weight')
plt.xlabel('Age')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(height, weight)
plt.title('Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(age, height, s=weight)
plt.title('Relationship between Age, Height, and Weight')
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(people, age)
plt.title('Age of Each Person')
plt.xlabel('Person')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(age, bins=5)  # Adjust the number of bins as needed
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
