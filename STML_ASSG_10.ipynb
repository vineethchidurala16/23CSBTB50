```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv('/content/breast_cancer_survival (5).csv')
d
```

```python
d.isnull().sum()
d.duplicated().sum()
d.drop_duplicates(inplace=True)
d.dropna(inplace=True)
d
```

```python
d['Gender']=d['Gender'].replace(['FEMALE','MALE'],[1,0])
d['Tumour_Stage']=d['Tumour_Stage'].replace(['I','II','III'],[0,1,2])
d['Histology']=d['Histology'].replace(['Infiltrating Ductal Carcinoma','Mucinous Carcinoma'],[1,2])
d['Histology']=d['Histology'].replace(['Infiltrating Lobular Carcinoma'],[0])
d['ER status']=d['ER status'].replace(['Positive','Negative'],[1,0])
d['PR status']=d['PR status'].replace(['Positive','Negative'],[1,0])
d['HER2 status']=d['HER2 status'].replace(['Positive','Negative'],[1,0])
d['Patient_Status']=d['Patient_Status'].replace(['Alive','Dead'],[1,0])
d['Surgery_type']=d['Surgery_type'].replace(['Lumpectomy','Modified Radical Mastectomy','Simple Mastectomy','Other'],[1,0,2,3])
d
```

```python
y=d['Patient_Status']
y
```

```python
x=d.drop(['Patient_Status','Date_of_Surgery','Date_of_Last_Visit'],axis=1)
x
```

```python
def diff_train_SVM(x, y, test_size, random_state):
    from sklearn.model_selection import train_test_split
    from sklearn.svm import SVC
    from sklearn.metrics import accuracy_score

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=random_state)
    model = SVC()
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)
    return accuracy
```

```python
acc1=diff_train_SVM(x, y, test_size=0.20, random_state=0)
```
